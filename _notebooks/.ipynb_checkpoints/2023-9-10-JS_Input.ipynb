```python
/*
  Whenever we go shopping, we have a list of stuff we want to buy.
  Create a program in which a user can store their name, the list of
  stuff they want, the day they plan to go shopping,their 
  phone number and a password. 

  Features:
  Pt 1:
  The program will ask if the user has created an account before.
  If so, it will ask for the name and then the password. Do not
  worry about whether or not they inputted something wrong.
  If not, ask them to create an account by asking for phone number,
  name and password

  Pt2:
  The program will ask what the user wants to buy. The user will type in a command if they are done, and the program will ask what day they plan to shop in. Then it will go back to login. Otherwise, it will add the item to the list and ask again.  

  Pt3:
  Keep on running the program until the user wants to leave the program
  You must include these cases:
  User switches accounts
  if the account exists: 
    User wants to add more stuff to the list
    User wants to change their email and phone number
*/
// var list = {item: "apple"};
// fruit = list["item"] //or list.item;

var list = {
    items: null,
    store: null,
    date: null
    };
    
    var shoppingList = []
    //userAccountList is used to store all of the user accounts for future
    //use and update
    var userAccountList = {}
    while(true){
      var signup = {}
      shoppingList = []
      var welcome = prompt("Welcome to the first and best virtual shopping list! If at anytime you want to leave, say the word 'done' and your receipt will be saved and a date will be added. But before we continue, do you have a account?                           (yes/no)")
      if(welcome == "yes"){
        var enterUsername = prompt("Please enter your username.")
        var enterPassword = prompt("Please enter your password")
        var userAccount = enterUsername
        //This checks if the username has been used and has an account
        // if(signup.username != null){
        //   var detectingAccount = prompt("We are detecting an account, would you like to be logged back in? (yes or no)")
        //   continue
        // }
        if(userAccountList[enterUsername] == undefined){
          console.log("That username doesn't exist. Please create an account")
          continue
        }
        signup = userAccountList[enterUsername]
        shoppingList = signup.shoppingList
        console.log(signup.password);
        while(enterPassword != signup.password){
          enterPassword = prompt("Please enter your CORRECT password")
        }
        console.log("Welcome back " + enterUsername)
          // var editSL = prompt("Would you like to edit your shopping list? (yes/no)") checks if the username has been used and has an account
        // if(signup.username != null){
        //   var detectingAccount = prompt("We are detecting an account, would you like to be logged back in? (yes or no)")
        //   continue
        // }
      }
      if(welcome == "no"){
        console.log("That's okay! Don't worry, creating a   account is a simple and free proccess!")
        var createUsername = prompt("First, you will need to enter a username.")
        var createPassword = prompt("Next, you have to create your password. Make sure that it is strong and secure.")
        var enterEmail2 = prompt("Please enter a pre-existing email that you own. This way, we can remind you by sending you a message.")
        var enterPhoneNumber = prompt("Lastly, make sure to enter in your phone number.")
        signup.username = createUsername
        signup.password = createPassword
        signup.email = enterEmail2
        signup.phoneNumber = enterPhoneNumber
    
        userAccountList[createUsername] = signup
      }
    var response = prompt("Now it's time to add items! When you want to stop, type 'done' and your receipt will be saved and a date will be added. Would you like to start? (yes/no)")
    // if(userInput == "yes"){
    //   var response = prompt("Great! What would you like to add?")
    // //   if(typeof response == "string"){
    // //     shoppingList.push(response)
    // //     console.log(response + " has been added to your shopping list!")
    // //   } 
    // // }
      if(response == "no"){
        console.log("Goodbye, have a nice day!")
        continue;
      }
    
      else{
        while(response != "done"){
          response = prompt("Great! What would you like to add?")
          shoppingList.push(response)
          console.log(response + " has been added to your shopping list.")
        }
      }
    
      console.log("Thanks for visiting! Wait for us to print out your list.")
      shoppingList.pop()
      let items = shoppingList
      let uniqueItems = new Set(items)
      console.log(uniqueItems)
      list.items = uniqueItems
      signup.shoppingList = shoppingList
    
    var enterStore = prompt("What store would you like to use this shopping list for?")
    var enterDate = prompt("What day would you like to use this list?(So that we can remind you!)")
    list.store = enterStore
    list.date = enterDate
    userAccountList[signup.username] = signup
    console.log("Your reminder will be sent to" + " " + signup.email + " " + "on" + " " + list.date + ", "  + " " + list.date + ", " + "until then, goodbye!")
    }
        
```
